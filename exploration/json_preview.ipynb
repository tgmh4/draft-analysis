```python
def preview_json(obj, n: int = 5):
    """
    Preview a JSON-like object in a DataFrame- or list-friendly format,
    similar to df.head() for quick inspection.

    Args:
        obj (dict | list | primitive): JSON object or list to preview.
        n (int, optional): Number of rows/items to display. Defaults to 5.

    Returns:
        pd.DataFrame | list | primitive: A formatted preview of the data.
    """
    # Case 1: List of dicts ‚Üí flatten into DataFrame
    if isinstance(obj, list) and obj and isinstance(obj[0], dict):
        return pd.json_normalize(obj).head(n)

    # Case 2: Plain list ‚Üí return first n elements
    if isinstance(obj, list):
        return obj[:n]

    # Case 3: Dict ‚Üí return as 2-column DataFrame
    if isinstance(obj, dict):
        return (
            pd.DataFrame(obj.items(), columns=["Key", "Value"])
            .head(n)
        )

    # Case 4: Anything else ‚Üí return directly
    return obj
```

### `teamfights`

Each element in the outer list is one fight/event:

	‚Ä¢	start: Game time (in seconds) when this fight started.
	‚Ä¢	end: Game time when the fight ended.
	‚Ä¢	last_death: Time of the last death in the fight (seconds).
	‚Ä¢	deaths: Total number of deaths that occurred during this fight.
	‚Ä¢	players: A list of per-player stats for this interval.

    üîπ Per-Player Structure

Each item inside players corresponds to one hero in the match.

Key Fields
	‚Ä¢	deaths_pos: Where deaths occurred (map coords). Here, hero died once at X=140.7, Y=128.5.
	‚Ä¢	ability_uses: Which spells they cast (e.g. rattletrap_power_cogs: 1).
	‚Ä¢	ability_targets: (Empty here) would log spell ‚Üí target mappings.
	‚Ä¢	item_uses: Items activated (e.g. magic_stick: 1).
	‚Ä¢	killed: Dictionary of units killed, often npc_dota_hero_*.
	‚Ä¢	deaths: Count of times this hero died in the fight.
	‚Ä¢	buybacks: Number of buybacks used.
	‚Ä¢	damage: Total damage dealt in this fight.
	‚Ä¢	healing: Total healing done in this fight.
	‚Ä¢	gold_delta: Net change in gold from this fight.
	‚Ä¢	xp_delta: Net change in XP.
	‚Ä¢	xp_start/xp_end: Their XP before/after this fight.


	‚Ä¢	The players array breaks down individual hero contributions: spells, items, kills, damage, XP, etc.
	‚Ä¢	Useful for fight analysis dashboards, ‚Äúimpact rating,‚Äù or training data for ML.

```python
def parse_teamfights(match_data):
    """
    Parse match_data['teamfights'] into a DataFrame,
    preserving nested dicts/lists as JSON columns.

    Parameters
    ----------
    match_data : dict
        The raw match JSON from the OpenDota API.

    Returns
    -------
    teamfights_df : pd.DataFrame
        One row per teamfight, with JSON columns preserved.
    """
    teamfights = match_data.get("teamfights", [])
    if not teamfights:
        return pd.DataFrame()

    # Convert directly to DataFrame (nested dicts stay as JSON/dict objects)
    teamfights_df = pd.DataFrame(teamfights)

    # Optional: ensure 'players' stays JSON, not expanded
    if "players" in teamfights_df.columns:
        teamfights_df["players"] = teamfights_df["players"].apply(lambda x: x if isinstance(x, list) else [])

    return teamfights_df
```

```python
teamfights_parsed = parse_teamfights(match_data)
teamfights_parsed
```

```python
def parse_objectives(match_data):
    """
    Parse the 'objectives' key from match_data into a DataFrame.
    Keeps scalar values as columns and leaves nested dicts/json-like
    fields intact for later exploration.
    """
    objectives = match_data.get("objectives", [])
    if not objectives:
        return pd.DataFrame()  # empty if no objectives
    
    # Normalize with pandas
    df = pd.json_normalize(objectives, sep="_")
    
    # Ensure nested structures remain visible but not exploded
    for col in df.columns:
        df[col] = df[col].apply(
            lambda x: json.dumps(x) if isinstance(x, (dict, list)) else x
        )
    
    return df
```

```python
objectives_parsed = parse_objectives(match_data)
objectives_parsed
```

```python
def parse_players(match_data):
    """
    Return players as a DataFrame:
    - Scalars become columns
    - Nested dicts/lists stay as JSON objects
    """
    players = match_data.get("players", [])
    if not players:
        return pd.DataFrame()

    # Build DataFrame, but don't normalize deeply
    df = pd.DataFrame(players)

    return df
```
